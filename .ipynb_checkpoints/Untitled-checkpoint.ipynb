```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import heapq
import random
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import tkinter as tk
```

```python
def parse_txt(file_path):
    graphs = []  
    current_matrix = []
    with open(file_path, 'r') as f:
        for line in f:
            line = line.strip()
            
            if line.startswith("Graph"):
                if current_matrix:
                    adj_matrix = np.array(current_matrix, dtype=int)
                    G = nx.from_numpy_array(adj_matrix)
                    for u, v in G.edges():
                        G[u][v]['weight'] = random.randint(1, 10) 
                    graphs.append(G)
                    current_matrix = [] 
            
            elif line:  
                    current_matrix.append([int(char) for char in line])
            
        if current_matrix:
            adj_matrix = np.array(current_matrix, dtype=int)
            G = nx.from_numpy_array(adj_matrix)
            for u, v in G.edges():
                G[u][v]['weight'] = random.randint(1, 10)
            graphs.append(G)
            
        return graphs
```

```python
def visualize_graphs(graphs):
    i = 1  
    for graph in graphs:
        plt.figure(figsize=(6, 6))
        
        pos = nx.spring_layout(graph, seed=42) 
        
        nx.draw(graph, pos, with_labels=True, node_color='pink', edge_color='gray')
        
        edge_labels = nx.get_edge_attributes(graph, 'weight')
        
        nx.draw_networkx_edge_labels(graph, pos, edge_labels=edge_labels)
        
        plt.title(f"Graph {i}")
        
        plt.show()
        i += 1
        
graphs = parse_txt("graph_output.txt")
visualize_graphs(graphs)
```

```python
def prim_mst(G):
    
    start_node = list(G.nodes())[0]  
    visited = set([start_node])
    
    edges = [
        (G[start_node][dest]["weight"], start_node, dest)
        for dest in G[start_node]
    ]
    heapq.heapify(edges)  
    mst = []  
    
    while edges and len(visited) < len(G.nodes):
        weight, src, dest = heapq.heappop(edges)
        if dest in visited:
            continue
        visited.add(dest)
        mst.append((weight, src, dest))
        
        
        for neighbor, edge_attr in G[dest].items():
            if neighbor not in visited:
                edge_weight = edge_attr["weight"]
                heapq.heappush(edges, (edge_weight, dest, neighbor))
    
    return mst
```

```python
i = 1  
for graph in graphs:
    mst_edges = prim_mst(graph)
    print(f"MST for Graph {i}:", mst_edges)
    i += 1  
```

```python
class MSTVisualizer:
    def __init__(self, master, graphs):
        self.master = master
        self.master.title("MST Step-by-Step Visualization")
        
        self.graphs = graphs
        self.current_graph_index = 0
        self.current_step = 0
        self.mst_edges = []  # Stores MST edges for the current graph
        self.graph = self.graphs[self.current_graph_index]
        
        self.fig, self.ax = plt.subplots(figsize=(6, 6))
        self.canvas = FigureCanvasTkAgg(self.fig, master=self.master)
        self.canvas.get_tk_widget().pack()
        
        self.next_button = tk.Button(master, text="Next Step", command=self.next_step)
        self.next_button.pack()
        
        self.init_graph()

    def init_graph(self):
        """Displays the initial graph without MST edges."""
        self.ax.clear()
        pos = nx.spring_layout(self.graph, seed=42) 
        nx.draw(self.graph, pos, ax=self.ax, with_labels=True, node_color='lightblue', edge_color='gray')
        edge_labels = nx.get_edge_attributes(self.graph, 'weight')
        nx.draw_networkx_edge_labels(self.graph, pos, edge_labels=edge_labels, ax=self.ax)
        
        self.canvas.draw()
        
        # Compute MST for the current graph
        self.mst_edges = prim_mst(self.graph)
        self.current_step = 0

    def next_step(self):
        """Highlights the next MST edge on the graph."""
        if self.current_step < len(self.mst_edges):
            weight, src, dest = self.mst_edges[self.current_step]
            self.ax.plot(*zip(*nx.spring_layout(self.graph, seed=42).values()), 'ro')  # To redraw nodes in red
            nx.draw_networkx_edges(self.graph, pos=nx.spring_layout(self.graph, seed=42), ax=self.ax,
                                   edgelist=[(src, dest)], edge_color='red', width=2)
            self.canvas.draw()
            self.current_step += 1

root = tk.Tk()
app = MSTVisualizer(root, graphs)
root.mainloop()
```
